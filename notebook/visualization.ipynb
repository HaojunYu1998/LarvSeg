```python
def pairwise_affinity_score_map(
    image,
    embedding,
    label,
    label_to_vis,
    dot_color="red",
    color_map = cv2.COLORMAP_VIRIDIS,
    interpolate=False,
    alpha=1.0
):
    # embedding = embedding.permute(0, 3, 1, 2)
    # if interpolate:
    h, w, _ = image.shape
    embedding = F.interpolate(embedding, size=(h, w), mode="bilinear", align_corners=True)
    # else:
    #     h, w = embedding.shape[-2:]
    #     image = F.interpolate(
    #         torch.from_numpy(image).permute(2,0,1)[None].cuda().float(), size=(h, w), mode="bilinear", align_corners=True
    #     )[0].permute(1,2,0).cpu().numpy()
    #     label = F.interpolate(
    #         torch.from_numpy(label)[None, None].cuda().float(), size=(h, w), mode="nearest"
    #     )[0,0].long().cpu().numpy()
    embedding = embedding.reshape(-1, h * w).permute(1, 0)
    image_vis = Visualizer(image.copy())
    mask = label == label_to_vis
    inds = np.nonzero(mask.reshape((-1,)))[0]
    inds = np.random.choice(inds, size=1)
    selected_embedding = embedding[torch.from_numpy(inds).cuda()]
    one_sim = F.cosine_similarity(
        selected_embedding[:, None, :], embedding[None, ...], dim=-1
    ).reshape(h, w)
    coord = np.stack([inds % w, inds // w], axis=1)[0]  # k,2
    one_sim = one_sim.cpu().numpy()
    heatmap = (one_sim - np.min(one_sim)) / (np.max(one_sim)-np.min(one_sim))
    heatmap = np.uint8(255 * heatmap)
    heatmap = cv2.cvtColor(
        cv2.applyColorMap(heatmap, color_map), cv2.COLOR_BGR2RGB
    )
    score_vis = Visualizer(image.copy())
    # draw heatmap
    score_vis.output.ax.imshow(heatmap, alpha=alpha)
    score_vis.draw_circle([coord[0], coord[1]], radius=5, color=dot_color)
    image_vis.draw_circle([coord[0], coord[1]], radius=5, color=dot_color)
    score_img = score_vis.get_output().get_image()
    image = image_vis.get_output().get_image()
    return image, score_img, coord
```

```python
def _coseg_score(embed, queue, cls_name, name, unique_label):
    fg = unique_label[cls_name.index(name)]
    bgs = [l for l in unique_label if l != fg]
    fg_embed = queue[fg].reshape(-1, 768)
    fg_embed = fg_embed / fg_embed.norm(dim=-1, keepdim=True)
    embed = embed / embed.norm(dim=-1, keepdim=True)
    cos_mat = embed @ fg_embed.T
    coseg_score = cos_mat.mean(dim=-1)
    if len(bgs) > 0:
        bg_embed = queue[bgs]
        bg_embed = bg_embed.reshape(-1, 768)
        bg_embed = bg_embed / bg_embed.norm(dim=-1, keepdim=True)
        cos_mat = embed @ bg_embed.T
        bg_coseg_score = cos_mat.reshape(len(embed), len(bgs), -1).mean(dim=-1)
    if len(bgs) > 0 and bg_coseg_score.numel() > 0:
        bg_coseg_score = bg_coseg_score.mean(dim=-1)
        coseg_score = coseg_score - bg_coseg_score
    return coseg_score
```
