```python
def merge_label(path, save_dir):
    from PIL import Image
    import numpy as np
    import os
    label = np.array((Image.open(path))).astype(np.uint16)
    unique_labels = np.unique(label)
    unique_labels = unique_labels[unique_labels != 65535]
    new_label = np.zeros_like(label) + 65535
    for l in unique_labels:
        new_label[label == int(l)] = ade_full_to_merged_id[l]
    Image.fromarray(new_label.astype(np.uint16)).save(
        os.path.join(save_dir, path.split("/")[-1]), format="TIFF"
    )
    # print(np.unique(new_label), np.unique(label))
```

```python
def get_wordnet_hyper(cat, offset):
    from nltk.corpus import wordnet as wn
    # synsets = wn.synsets(cat, pos=wn.NOUN)
    synset = wn.synset_from_pos_and_offset('n',int(offset[1:]))
    hyper_synset = synset.hypernyms()
    return [str(s.name())  for syn in hyper_synset for s in syn.lemmas()]
```
